```python
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset = pd.get_dummies(dataset)
```

```python
independant = dataset[['age', 'sex_male', 'bmi', 'children', 'smoker_yes']]
dependant = dataset[['charges']]
```

```python
dataset.columns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independant, dependant, test_size = 0.30, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid = { 'criterion' :['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],'n_estimators':[100, 150, 200,], 'max_features':['auto', 'sqrt', 'log2'] }
regressor = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose = 3, n_jobs = -1)
regressor.fit(x_train, y_train)
```

```python
result = regressor.cv_results_
table = pd.DataFrame.from_dict(result)
regressor.best_score_
```

```python
predict = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, predict)
print(f'The model score  is {r_score}')
```
