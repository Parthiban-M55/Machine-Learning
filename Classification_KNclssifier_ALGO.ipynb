```python

from sklearn.preprocessing import  LabelEncoder
lable_encoder = LabelEncoder()
nominal_cols = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane']

for cols in nominal_cols:
    dataset[cols] = lable_encoder.fit_transform(dataset[cols])
ordinal_mapping = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
dataset['sg'] = dataset['sg'].map(ordinal_mapping)

dataset['sg'].value_counts()
```

```python
dependent_variable = dataset['classification']
independent_variable = dataset.drop('classification', axis=1)

independent_variable
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent_variable,dependent_variable, test_size=0.3, random_state=0)

x_train['su'].value_counts()


```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
# from sklearn.svm import SVC
# from sklearn.tree import DecisionTreeClassifier
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.naive_bayes import GaussianNB

grid_params = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan','minkowski'],
    'p': [1, 2]
}
model = KNeighborsClassifier()
grid_search = GridSearchCV(estimator=model, param_grid=grid_params, refit=True, verbose=3,n_jobs=-1 ,scoring='f1_weighted')
grid_search.fit(x_train, y_train)
y_pred = grid_search.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cls_report = classification_report(y_test, y_pred)
print(cls_report)
```
